```python
# Importing recquired packages
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
import time
import pandas as pd
```

```python
# Locating the driver
driver_location = "/Users/camilleepitalon/chromedriver"
driver = webdriver.Chrome(driver_location)

# Activating the driver on frimm landing page
driver.get('https://www.facebook.com/events/476096280549566/?acontext=%7B%22event_action_history%22%3A[%7B%22mechanism%22%3A%22search_results%22%2C%22surface%22%3A%22search%22%7D]%2C%22ref_notif_type%22%3Anull%7D')
time.sleep(3)

# Closing the cookies pop-up banner
driver.find_element_by_xpath("//button[text()='Autoriser les cookies essentiels et optionnels']").click()
time.sleep(1)
driver.find_element_by_xpath("//input[@type='email']").send_keys('cameptlnfa@outlook.fr')
driver.find_element_by_xpath("//input[@type='password']").send_keys('Mockingbird1997@fa', Keys.ENTER)
```

```python
participants = []

name_elements = driver.find_elements_by_xpath("//div[@class='qzhwtbm6 knvmm38d']/span/div/object[@type='nested/pressable']/a/span")
for name in name_elements:
    participants.append(name.text)

df_participants = pd.DataFrame(participants, columns=['Nom'])
df_participants['Statut'] = 'Participe'
df_participants
```

```python
peutetre = []

name_elements = driver.find_elements_by_xpath("//div[@class='qzhwtbm6 knvmm38d']/span/div/object[@type='nested/pressable']/a/span")
for name in name_elements:
    peutetre.append(name.text)

df_peutetre = pd.DataFrame(peutetre, columns=['Nom'])
df_peutetre['Statut'] = 'Peut-être'
df_peutetre
```

```python
invités = []

name_elements = driver.find_elements_by_xpath("//div[@class='qzhwtbm6 knvmm38d']/span/div/object[@type='nested/pressable']/a/span")
for name in name_elements:
    invités.append(name.text)

df_invités = pd.DataFrame(invités, columns=['Nom'])
df_invités['Statut'] = 'Invités'
df_invités
```

```python
ne_peut_pas = []

name_elements = driver.find_elements_by_xpath("//div[@class='qzhwtbm6 knvmm38d']/span/div/object[@type='nested/pressable']/a/span")
for name in name_elements:
    ne_peut_pas.append(name.text)

df_ne_peut_pas = pd.DataFrame(ne_peut_pas, columns=['Nom'])
df_ne_peut_pas['Statut'] = 'Ne peut pas'
df_ne_peut_pas
```

```python
df_facebook = pd.concat([df_participants, df_peutetre, df_invités, df_ne_peut_pas])
df_facebook
```

```python
# Exporting the dataframe to a CSV file
df_facebook.to_excel(r'/Users/camilleepitalon/Desktop/df_facebook.xlsx', encoding='utf-8-sig', index = False)
```
