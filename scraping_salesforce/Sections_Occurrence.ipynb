```python
# Importing needed packages
from bs4 import BeautifulSoup
from selenium import webdriver
import time
from collections import Counter
```

```python
# Locating the driver
driver_location = "/Users/macbook16decamille/chromedriver"    # To download the driver, it's here > https://chromedriver.chromium.org/
driver = webdriver.Chrome(driver_location)

# Activating the driver on the first Standard Object page
driver.get("https://developer.salesforce.com/docs/atlas.en-us.object_reference.meta/object_reference/sforce_api_objects_acceptedeventrelation.htm")
```

```python
# Closing Cookies Preference popups on the page
time.sleep(2)
driver.find_element_by_css_selector("button#onetrust-accept-btn-handler").click()
time.sleep(2)
driver.find_element_by_css_selector("img.close-icon").click()

# Creating the list of Objects
object_list=[]
for el in driver.find_elements_by_css_selector("span.ng-binding.ng-scope") :
    object_list.append(el.text)
```

```python
# Formatting the list of Objects

# Keeping Standard Objects only
idx = object_list.index("AcceptedEventRelation")
idx2 = object_list.index("WorkTypeGroupMember")
object_list = object_list[idx:idx2]

# Lowering the objects names
for idx,obj in enumerate(object_list) : 
    object_list[idx] = obj.lower()
print(object_list)
```

```python
# Extracting all the sections' titles from all standard objects pages

titles_list = []
error_pages = []

# Extracting the titles of the first page thanks to Beautiful Soup
soup = BeautifulSoup(driver.page_source,'lxml')
for title in soup.select("h2.helpHead2") : 
    titles_list.append(title.text)

# Extracting the titles of all other pages thanks to Beautiful Soup
for url in object_list[1:] :
    driver.get("https://developer.salesforce.com/docs/atlas.en-us.object_reference.meta/object_reference/sforce_api_objects_"+url+".htm")
    time.sleep(2)
    soup=BeautifulSoup(driver.page_source,'lxml')
    list_titles_html = soup.select("h2.helpHead2")
    # To avoid Index Errors, an if condition is needed
    if len(list_titles_html) > 0:
        for title in list_titles_html : 
            titles_list.append(title.text)
    # Gathering all the pages without title in this list
    else : error_pages.append(url)
```
